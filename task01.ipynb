```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("data(2).csv")
data
```

```python
#removing duplicate data entries
data.drop_duplicates()
data
```

```python
#dropping irrelevant columns
data2=data.iloc[:, [i for i in range(28, 53)]]
data=data.drop(columns=data2.columns, errors='ignore')
data
```

```python
#performing data type conversions
data['Release Claue']=data['Release Clause'].str.replace('[^0-9.]', '', regex=True)
data
```

```python
data['Release Clause']=pd.to_numeric(data['Release Clause'])
data
```

```python
#analysing missing values
missing_values=data.isna()
data=data.fillna(data.mean())
data
```

```python
##Visulalisation
#bar plot
plt.bar(data['Age'], data['Potential'])

spacing = 5  # Define the spacing between ticks

xticks = np.arange(0, max(data['Age']) + spacing, spacing)  # Generate the tick positions
xtick_labels = [str(x) for x in xticks]  # Convert positions to string labels
plt.xticks(xticks, xtick_labels)

spacing = 5  # Define the spacing between ticks
yticks = np.arange(0, max(data['Potential']) + spacing, spacing)  # Generate the tick positions
ytick_labels = [str(y) for y in yticks]  # Convert positions to string labels
plt.yticks(yticks, ytick_labels)

plt.show()
```

```python
#lineplot
y=data['Age'].value_counts().sort_index()

plt.plot(y.index, y.values, marker='o', linestyle='-')
plt.xlabel("Age")
plt.ylabel("Age frequency")

plt.title('Line Plot')
```

```python
#Pie chart

y=data['Preferred Foot'].value_counts()
plt.pie(x=y, labels=y.index)
plt.show()
```
